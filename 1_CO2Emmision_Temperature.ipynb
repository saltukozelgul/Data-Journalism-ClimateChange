```python

df = pd.read_csv('C:/Users/berke/OneDrive/Masaüstü/emission.csv', index_col='Country Name')

df1=df.transpose()
df1.index=pd.to_datetime(df1.index).year
df1.head()

```

```python
# Reset the index of the df data frame
df.reset_index(level=0, inplace=True)
```

```python
# Melt the df data frame and sort the resulting data frame by year
df2 = df.melt(id_vars=['Country Name'], var_name='Year').sort_values(by=['Year'])
```

```python
# Create a choropleth map of CO2 emissions by country and year
px.choropleth(data_frame=df2[df2['Country Name']!='World'], locationmode='country names',locations='Country Name', color='value', animation_frame='Year', title='CO2 Emission by Countries by Year',color_continuous_scale=px.colors.sequential.RdBu_r, range_color=(200, 0))
```

```python
# Create a bar plot of CO2 emissions by country over time
race_plot = barplot(df=df2[df2['Country Name']!='World'], item_column='Country Name', value_column='value', time_column='Year', top_entries=10)

# Plot the bar plot and customize the labels and title
race_plot.plot(title='Carbon Emission Race by Country (Since 1960 to 2018)', value_label='Value', item_label='Top 10 Countries')
```

```python
# Get the top 10 countries by CO2 emissions
country_col = df2[df2['Country Name']!='World'].groupby(by=['Country Name']).max().sort_values(by=['value'], ascending=False).head(10).index.to_list()

# Add India and China to the list of countries
country_col.extend(['India','China'])
```

```python
# Calculate the mean CO2 emission for the selected countries
sample_mean = df2[df2['Country Name'].isin(country_col)]['value'].mean()

# Calculate the mean CO2 emission for the entire population
pop_mean = df2['value'].mean()
```

```python
# Create a figure with a specified size
plt.figure(figsize=(15,8))

# Create a kernel density plot of the CO2 emissions for the selected countries
ax = sns.kdeplot(data=df2[df2['Country Name'].isin(country_col)], x=df2['value'], hue='Country Name', fill=True, linewidth=3, edgecolor='white', alpha=0.7)

# Add a vertical line to indicate the sample mean
ax = plt.axvline(x=sample_mean)

# Add a vertical line to indicate the population mean
ax = plt.axvline(x=pop_mean)

# Add a text label to indicate the sample mean
ax = plt.text(x=sample_mean, y=0.015, s='Sample mean')

# Add a text label to indicate the population mean
ax = plt.text(x=pop_mean, y=0.015, s='Population mean')

# Set the x-axis limits
plt.xlim(-20,60)

# Set the y-axis limits
plt.ylim(0.00, 0.04)
```
