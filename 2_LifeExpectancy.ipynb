```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# load dataset
df = pd.read_csv('Datasets/life_expectancy_data.csv')

# only select data for years 2000-2015
df = df[(df['Year'] >= 2000) & (df['Year'] <= 2015)]

# calculate average life expectancy
df = df.groupby('Country')['Life expectancy '].mean().reset_index()

# Sort "Life expectancy" column from largest to smallest
df = df.sort_values('Life expectancy ', ascending=False)

# select first 10 lines
df = df.head(10)

# print results
print(df)

# create scatter plot graph
plt.ylim(0,100)
plt.scatter(df['Country'], df['Life expectancy '], c=range(len(df)))
plt.xticks(rotation=90)  # print labels on x-axis vertically
plt.xlabel('Top 10 Country')
plt.ylabel('Average Life Expectancy (2000-2015)')
plt.title('Average Life Expectancy (2000-2015)')
plt.show()

# create bar graph
plt.ylim(40,90)
plt.bar(df['Country'], df['Life expectancy '], color='pink')
plt.xticks(rotation=90)  # print labels on x-axis vertically
plt.xlabel('Top 10 Country')
plt.ylabel('Average Life Expectancy (2000-2015)')
plt.title('Average Life Expectancy (2000-2015)')
plt.show()


# In this piece of code, the average life expectancy for each country between the years 2000-2015 is calculated
# and the 10 countries with the longest average life expectancy are printed on the screen.
# In addition, a scatter plot graph was drawn showing these data.
# In addition, a bar graph was drawn showing these data.
```

```python
# load dataset
df = pd.read_csv('Datasets/life_expectancy_data.csv')

# Let's look at the data from 2000-2015
filtered_df = df[(df['Year'] >= 2000) & (df['Year'] <= 2015)]

# Calculate the average life expectancy for each country
grouped_df = filtered_df.groupby('Country')['Life expectancy '].mean().sort_values()

# Print the average life expectancy to the screen
print(grouped_df)

# Load average life expectancy into a data frame
data = grouped_df.reset_index()
data.columns = ['Country', 'Life expectancy']

# Create scatter graph
fig = px.scatter(data, x='Country', y='Life expectancy', title='Average Life Expectancy (2000-2015)')
fig.show()
```

```python
# Create a line graph
fig = px.line(data, x='Country', y='Life expectancy', title='Average Life Expectancy (2000-2015)')
fig.show()
```

```python
# Create an area graph
fig = px.area(data, x='Country', y='Life expectancy', title='Average Life Expectancy (2000-2015)')
fig.show()
```

```python
# Create a choropleth map
fig = px.choropleth(data, locations='Country', locationmode='country names', color='Life expectancy',
                    title='Average Life Expectancy (2000-2015)',color_continuous_scale='Purp')
fig.show()
```
